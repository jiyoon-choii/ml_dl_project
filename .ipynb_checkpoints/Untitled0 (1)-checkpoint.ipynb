### 링크 : https://en.wikipedia.org/wiki/BTS_(band)

```python
target_url = 'https://en.wikipedia.org/wiki/BTS_(band)'
target_url
```

```python
import urllib.request as req 
from bs4 import BeautifulSoup
```

```python
res = req.urlopen(target_url)
soup = BeautifulSoup(res, 'html5lib')
```

```python
#type(soup), soup
```

```python
# 대상 태그들을 모은다 
ps = soup.select('#mw-content-text p')
len(ps), type(ps)
```

```python
# ps에서 한 개 뽑아서 텍스트를 출력 
tmp = list()

for p in ps :
  # print('[%s]' % p.text) 
  tmp.append('[%s]' % p.text.strip())

tmp[:3], tmp[-3:]
   
  # 79개를 다 뽑아서 리스트에 담아라
```

```python
# 목적 - 텍스트 여러개를 리스트에 모은다
# 리스트 내포를 이용하여 79개의 p태그에서 텍스트 추출하여 리스트에 구성원으로 추가하시오 
tmp = [ p.text.strip() for p in ps ] ##  오 새로운 식 !!! 
tmp[:10]
```

```python
# 리스트에 존재하는 문자열들을 한개의 문자열로 통합 
src_txt = ' '.join(tmp)
len(src_txt), src_txt[:100]
```

- 영어는 소문자로 통일 (lower) 
- 영어를 제외하고는 모두 제거 (정규식) 
- 띄어쓰기 기준으로 정도는 남기고 싶다(가독성 올라감) (정규식 중에 공백은 두자) 

```python
# 정규식 
import re 

# 정규식, 알파벳만 남겨라 -> 
p = re.compile('[^a-zA-Z ]*') # Z 뒤에 한 칸 띄움으로써 공백 만듬(제일 쉬운 방법, 조금 안 맞을 수 있음 )
tmp = p.sub( '' , src_txt )
tmp[:100]
```

### 알파벳 빈도 계산 

```python
# 텍스트 파일을 읽어서 => 알파벳만 남긴다 
# 최종 변수명 = text

# 텍스트 파일을 읽어서(I/O) 
with open( file_list[0], encoding='utf-8' ) as f :
  # 읽어서
  text = f.read().lower()
  # 알파벳만 남기고 or 알파벳, 띄어쓰기까지 남기고 
     
  p = re.compile('[^a-z]*') 
  text = p.sub( '' , text ) # 일치되는 표현만 (치환)제거시키겠다 
  text 
  print( text[:100] )
  pass  

```

```python
# 알파벳 별로 카운트를 담을 그릇 (한개의 이름로 26개의 정보를 담는 그릇)
# 파이썬의 리스트를 사용한다. 수정가능하고, 순서가 존재하므로 적합하다 
# a-<z 까지는 순서가 존재하므로 (ASCII)
```

```python
[0]*26
```

```python
# text에서 한글자씩 읽어서 -> a면 counts[0]+=1
# z -> counts[25] += 1 -> counts[z-a]+=1 
# t  -> counts[?] += 1 -> counts[t-a] += 1 
# ord() 특정 문자의 아스키코드값을 리턴 
# a-z 까지는 연속적으로 아스킥값이 배치되어 있다
# 따라서 특정 문자의 위치는 ord('소문자)-ord('a') 
ord('a'), ord('z'), (ord('z')-ord('a'))
```

```python
counts = [ 0 for n in range(26) ]
ASCII_A = ord('a')
ASCII_A
```

```python
text
```

```python
for ch in text : 
  # print( ch )
  # 한개 뽑힌 문자의 아스키값 ord(ch)
  print(ord(ch))
  counts[ ord(ch)-ASCII_A ] += 1
  # break 
print( counts )
```

- 데이터를 들여다보니, 값의 편차가 크다 (최대, 최소 등으로 확인 )
- 데이터는 빈도를 표현함으로, 최대 최소의 의미보다는 총 빈도에서 특정 알파벳이 얼마나 등장했는가가 중요하다 => 이것을 기준으로 정규화처리 
- 데이터 준비 단계에서는 데이터의 품질 향상을 목표로 하고, 주 수단으로 정한다 

```python
# map 함수를 써라 
# list( map( 함수, 데이터덩어리 )
# 함수는 map이 자동으로 멤버수만큼 호출한다
# x : 멤버 한개 한개 
def checkNorm( x ) :
  # 알파벳별빈도/총빈도(=총알파벳문자열개수)
  return x / len(text) 
frequences = list( map( checkNorm, counts ) )
print( sum(frequences), frequences )
```

### 위의 구성 내용을 한칸에 넣어서 함수화 시킨다 

```python
import os,re

def parsingLangFreq( file_path ) : 
  # 언어코드 획득  
  name = os.path.basename( file_path )
  p = re.compile( '^[a-z]{2}' )
  lang_code = p.match( name ).group()
  lang_code 

  # 단어의 빈도 
  with open( file_path, encoding='utf-8' ) as f :  
    text = f.read().lower()     
    p = re.compile('[^a-z]*') 
    text = p.sub( '' , text ) 

  counts = [ 0 for n in range(26) ]
  ASCII_A = ord('a')
  for ch in text : 
    counts[(ord(ch)-ASCII_A)] += 1

  frequences = list( map( checkNorm, counts ) )

  # 결과 리턴 
  return( lang_code, frequences )

print( parsingLangFreq( file_list[0]) )
```

```python
file_list[0]
```

```python

# 훈련용 데이터, 테스트용 데이터 로드
# 중간 경로가 train or test 
def load_data( path='train' ) :
  # 파일 경로 획득 
  file_lists = glob.glob('./data/{}/*.txt'.format(path) )
  # 정답을 담을 리스트
  labels = list()
  freqs = list()
  # 파일 경로를 한 개씩 뽑아서
  for f in file_lists : 
    # 파일 1개당, 정답, 빈도수 리스트 리턴 
    lang, freq = parsingLangFreq( f )
    labels.append(lang)
    freqs.append(freq)

  return { 'labels' : labels, 'freqs':freqs }

train_data = load_data('train')
test_data = load_data('test')


```

```python
# 덤프 (현 단계 결과물을 저장)
import json

with open('./data/labels_freqs_data.json', 'w') as f : 
  # json 구조로 저장
  json.dump( [train_data, test_data], f )
```
